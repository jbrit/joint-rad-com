## Communication Rate vs SNR<sub>COM</sub> for C<sub>OFDM</sub>
$ C_{OFDM} = (\frac{T}{T + T_{cp}})log_2(1 + SNR_{com})$

$ T_{cp} = \frac{1}{4} T$

$\therefore C_{OFDM} = 0.8 log_2(1 + SNR_{com}) $

```python
from matplotlib import pyplot as plt
import numpy as np
import math

cofdm_from_snr_com = lambda snr_com: 0.8 * np.log2(1+10**(snr_com/10))
fig, ax = plt.subplots()
snr = np.linspace(0.0, 50.0, 26)
print(snr)
plt.plot(snr, cofdm_from_snr_com(snr), color='blue', marker='s', markerfacecolor='None', label=r'$C_{OFDM}$')
plt.legend()
plt.grid(color = 'grey', linestyle = '--', linewidth = 0.5)
plt.grid(which='minor', color = 'grey', linestyle = '--', linewidth = 0.25)
ax.minorticks_on()
ax.set(title=r"$SNR_{com}$ vs Communication Rate")
plt.xlim(xmin=0, xmax=50)
plt.ylim(ymin=0, ymax=15)
plt.show()
```
